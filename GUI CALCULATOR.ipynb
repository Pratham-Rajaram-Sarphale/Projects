```python
#GUI CALCULATOR
from tkinter import *
def click(event):
    global scvalue
    text=event.widget.cget("text")
    if text=="=":
        if scvalue.get().isdigit():
            value=int(scvalue.get())
        else:
            value=eval(screen.get())
        scvalue.set(value)
        screen.update()
    elif text=="AC":
        scvalue.set(" ")
        screen.update()
    else:
        scvalue.set(scvalue.get()+text)
        screen.update()
win=Tk()
win.wm_iconbitmap("Dakirby309-Simply-Styled-Calculator.ico")
win.geometry('600x700')
win.maxsize(600,700)
win.minsize(600,700)
win.config(bg='gray25')
win.title("PRS-Calculator")
scvalue=StringVar()
scvalue.set("")
f=Frame(win,padx=20,pady=20)
screen =Entry(f,textvar=scvalue, font="lucida 50 bold",bg="DarkSeaGreen4",borderwidth=5)
screen.pack(fill=X,padx=20,pady=15)
f.pack()
option1=['7','8','9','+']
option2=['4','5','6','-']
option3=['1','2','3','*']
option4=['0','.','AC','=','/']
f=Frame(win,bg='gray25',padx=30,pady=10)
for i in option1:
    b=Button(f,text=i,padx=24,pady=10,font="lucida 25 bold",bg='gray90')
    b.pack(side=LEFT,padx=10,pady=10)
    b.bind("<Button-1>",click)
f.pack()
f=Frame(win,bg='gray25',padx=30,pady=10)
for i in option2:
    b=Button(f,text=i,padx=25,pady=10,font="lucida 25 bold",bg='gray90')
    b.pack(side=LEFT,padx=10,pady=10)
    b.bind("<Button-1>",click)
f.pack()
f=Frame(win,bg='gray25',padx=30,pady=10)
for i in option3:
    b=Button(f,text=i,padx=25,pady=10,font="lucida 25 bold",bg='gray90')
    b.pack(side=LEFT,padx=10,pady=10)
    b.bind("<Button-1>",click)
f.pack()
f=Frame(win,bg='gray25',padx=30,pady=10)
for i in option4:
    b=Button(f,text=i,padx=11,pady=10,font="lucida 25 bold",bg='gray90')
    b.pack(side=LEFT,padx=10,pady=10)
    b.bind("<Button-1>",click)
f.pack()
win.mainloop()
```
